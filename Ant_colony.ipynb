```python
def Generate_Distance_Matrix(data):
    leng = data.shape[0]
    print(leng)
    dis = [[0 for i in range(leng)] for j in range(leng)] 
    for i in range(leng):
        for j in range(leng):
            dis[i][j] = np.sqrt((data.loc[i,'x'] - data.loc[j,'x'])**2 + (data.loc[i,'y'] - data.loc[j,'y'])**2)
    return dis
```
